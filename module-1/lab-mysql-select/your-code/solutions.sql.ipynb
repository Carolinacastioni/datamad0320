```python
import pandas as pd
```

```python
challenge_1 = pd.read_sql_query("""
SELECT 
	au.au_id, au.au_lname, au.au_fname, t.title, pub.pub_name
	FROM authors as au
	LEFT JOIN titleauthor as tta ON au.au_id = tta.au_id
	LEFT JOIN titles as t ON tta.title_id = t.title_id
	LEFT JOIN publishers as pub ON t.pub_id = pub.pub_id
    """, engine)
```

```python
challenge_1
```

```python
challenge_2 = pd.read_sql_query("""
SELECT 
	au.au_id, au.au_lname, au.au_fname, pub.pub_name, count(t.title) AS total
	FROM authors as au
	LEFT JOIN titleauthor as tta ON au.au_id = tta.au_id
	LEFT JOIN titles as t ON tta.title_id = t.title_id
	LEFT JOIN publishers as pub ON t.pub_id = pub.pub_id
	GROUP BY au.au_id, pub.pub_name
	ORDER BY au.au_fname
    """, engine)
```

```python
challenge_2
```

```python
challenge_3 = pd.read_sql_query("""
SELECT 
	au.au_id, au.au_lname, au.au_fname, sum(sl.qty) AS Total
	FROM authors as au
	LEFT JOIN titleauthor as tta ON au.au_id = tta.au_id
	LEFT JOIN sales as sl ON tta.title_id = sl.title_id
	GROUP BY au.au_id
	ORDER BY Total DESC
	LIMIT 3
    """, engine)
```

```python
challenge_3
```
